# Image Sharpening

In some images, the brightness and contrast may be satisfactory, yet the image still appears slightly **blurred**. In such cases, the edges are not sufficiently sharp, making it difficult to distinguish different objects and fine details.

For example:

- The branches of a tree may not be clearly distinguishable.
- The texture of the grass may appear smooth and uniform.
- Details of mountains or rocks may be difficult to observe.

In these situations, the problem is often caused by **weak or blurred edges**. Image **sharpening** enhances the visibility of these edges, making the image appear clearer and more detailed.


# What is Image Sharpening?

**Image Sharpening** is the process of enhancing the visibility of edges by increasing the intensity contrast around them.

As a result:

- The boundaries between objects become more distinct.
- Fine details become easier to observe.
- The image appears clearer and sharper.

It is important to note that **sharpening does not add new information to an image**. Instead, it increases the **local contrast** around edges, making existing details more noticeable and improving the perceived sharpness of the image.

# Sharpening Filters

One of the simplest approaches to **image sharpening** is the use of **convolution kernels** that emphasize edges and fine details.

Two of the most commonly used sharpening kernels are:

### **Sharpening Kernel 1**

```text
 0  -1   0
-1   5  -1
 0  -1   0
```

### **Sharpening Kernel 2**

```text
-1  -1  -1
-1   9  -1
-1  -1  -1
```

Both kernels sharpen the image by **increasing the contribution of the center pixel while subtracting the influence of its neighboring pixels**. This enhances local intensity differences around edges, making object boundaries and fine details appear clearer and more distinct.

The second kernel generally produces **stronger sharpening** because it considers all eight neighboring pixels, resulting in a more pronounced enhancement of edges.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
````

````python
images= ['old.jpeg','lowcontrast.jpg']

sharpen_filter1 = np.array([[ 0, -1,  0],
                            [-1,  5, -1],
                            [ 0, -1,  0]])


sharpen_filter2 = np.array([[ -1, -1,  -1],
                            [ -1,  9,  -1],
                            [ -1, -1,  -1]])


for image in images:
    img = cv2.cvtColor(cv2.imread(image),cv2.COLOR_BGR2RGB)
    img_sharp1 = cv2.filter2D(img,-1,sharpen_filter1)
    img_sharp2 = cv2.filter2D(img,-1,sharpen_filter2)

    plt.figure(figsize=(12,6))

    plt.subplot(1,3,1)
    plt.imshow(img)
    plt.title('img')
    plt.axis('off')


    plt.subplot(1,3,2)
    plt.imshow(img_sharp1)
    plt.title('img_sharp1')
    plt.axis('off')

    plt.subplot(1,3,3)
    plt.imshow(img_sharp2)
    plt.title('img_sharp2')
    plt.axis('off')

    plt.tight_layout()
    plt.show()





````
