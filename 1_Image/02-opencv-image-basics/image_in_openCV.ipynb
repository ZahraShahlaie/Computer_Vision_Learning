# GrayScale

change BGR bird image  and RGB bird image to grayscale image for show diffrence 


```python
img_bird_bgr_gray = cv2.cvtColor(img_bird , cv2.COLOR_BGR2GRAY)
img_bird_rgb_gray = cv2.cvtColor(img_bird_rgb , cv2.COLOR_RGB2GRAY)


plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img_bird_bgr_gray , cmap='gray')
plt.title("BGR→ Gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_bird_rgb_gray, cmap='gray')
plt.title("RGB → Gray")
plt.axis('off')

plt.tight_layout()
plt.show()
```

show info of gray image 

```python
print("img_bird_rgb_gray shape :",img_bird_rgb_gray.shape)
print("img_bird_rgb_gray max pixel :",img_bird_rgb_gray.max())
print("img_bird_rgb_gray min pixel :",img_bird_rgb_gray.min())
print("img_bird_rgb_gray type :",type(img_bird_rgb_gray))
```

we see that is (h,w) shape in gray image but im color image we had (h , w , d) shape
