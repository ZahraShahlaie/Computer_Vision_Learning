## 📊 Image Histogram and Contrast Analysis

#### 📌 Overview

A histogram shows the distribution of pixel intensity values in an image.

By comparing histograms, we can understand:

- Brightness distribution
- Image contrast
- Dynamic range
- Exposure differences between images

In this example, we compare two images of Milad Tower and calculate their **minimum intensity**, **maximum intensity**, and **Michelson contrast**.

---

#### 📌 Step 1: Import Libraries

```python
import cv2
import numpy as numpy
import matplotlib.pyplot as plt
```

#### 📌 Step 2: Create a Function to Compute Image Information

The following function **converts an RGB image to grayscale** and computes:

- Minimum intensity
- Maximum intensity
- Michelson Contrast



```python
def info_img(img):
    # Convert RGB image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    # Find minimum and maximum intensity
    min_val, max_val, _, _ = cv2.minMaxLoc(gray)

    # Michelson Contrast
    contrast = (max_val - min_val) / (max_val + min_val)

    return (
        f"Min Intensity : {min_val}\n"
        f"Max Intensity : {max_val}\n"
        f"Contrast : {contrast:.3f}"
    )
```

#### 📌 Step 3: Display Images and Their Histograms

```python
# Read Images

img_milad_1 = cv2.imread("miladtower1.jpg")
img_milad_2 = cv2.imread("miladtower2.jpg")

# Since OpenCV reads images in BGR format, convert them to RGB before displaying with Matplotlib.

img_milad_1_rgb = cv2.cvtColor(img_milad_1, cv2.COLOR_BGR2RGB)
img_milad_2_rgb = cv2.cvtColor(img_milad_2, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,8))

# ---------------- Image 1 ----------------
plt.subplot(2,2,1)
plt.imshow(img_milad_1_rgb)
plt.title("Milad Tower 1")
plt.axis("off")

# ---------------- Image 2 ----------------
plt.subplot(2,2,2)
plt.imshow(img_milad_2_rgb)
plt.title("Milad Tower 2")
plt.axis("off")

# ---------------- Histogram 1 ----------------
plt.subplot(2,2,3)

plt.hist(img_milad_1_rgb.ravel(), bins=100)

plt.title("Histogram - Image 1")

plt.xlabel(info_img(img_milad_1_rgb))

# ---------------- Histogram 2 ----------------
plt.subplot(2,2,4)

plt.hist(img_milad_2_rgb.ravel(), bins=100)

plt.title("Histogram - Image 2")

plt.xlabel(info_img(img_milad_2_rgb))

plt.tight_layout()
plt.show()
```
